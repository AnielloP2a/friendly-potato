```python
!pip install yfinance requests beautifulsoup4
import yfinance as yf

# Estrarre i dati delle azioni di Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")  # Ottieni l'intera cronologia
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup

# URL del sito web da cui fare scraping (da adattare in base alle tue esigenze)
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Aggiungi header per simulare un browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Estrarre i dati di fatturato (modifica in base alla struttura HTML della pagina)
# Check if there are any tables before trying to access them
if len(soup.find_all('table')) > 1:
    revenue_table = soup.find_all('table')[1]
    tesla_revenue = []

    for row in revenue_table.find_all('tr')[1:]:
        columns = row.find_all('td')
        # Check if there are enough columns before trying to access them
        if len(columns) >= 2:
            date = columns[0].text.strip()
            revenue = columns[1].text.strip()
            tesla_revenue.append((date, revenue))

    print(tesla_revenue[:5])  # Visualizza i primi 5 risultati
else:
    print("Could not find the expected table on the page.")
```

```python
import matplotlib.pyplot as plt

# Creazione del grafico per le azioni
plt.figure(figsize=(14, 7))
plt.plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock Price')
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Tesla Stock Price History")
plt.legend()
plt.show()

# Creazione del grafico per il fatturato
dates, revenues = zip(*tesla_revenue)
plt.figure(figsize=(14, 7))
plt.plot(dates, revenues, label='Tesla Revenue')
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Tesla Revenue Over Time")
plt.legend()
plt.show()
```
